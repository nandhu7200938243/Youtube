```python
import pandas as pd
df=pd.read_csv(r"C:\Users\ngkna\OneDrive\Documents\Data Science\Cognify Data Science\Dataset .csv")
df.head()
```

```python
df["Has Table booking"]=df["Has Table booking"].map({'Yes':1,'No':0})
```

```python
df["Has Online delivery"]=df["Has Online delivery"].map({'Yes':1,'No':0})
```

```python
df.head()
```

```python
x = df[["Price range","Votes","Longitude","Latitude","Has Table booking","Has Online delivery"]]
y = df["Aggregate rating"]
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import r2_score, mean_squared_error
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8,random_state=10)
```

```python
model1=LinearRegression()
model1.fit(x_train,y_train)
y_pred_1=model1.predict(x_test)
r2_1=r2_score(y_test,y_pred_1)
rmse_1=np.sqrt(mean_squared_error(y_test,y_pred_1))

print("Linear Regression R2:", r2_1)
print("Linear Regression RMSE:", rmse_1)
```

```python
model2=DecisionTreeRegressor(random_state=10)
model2.fit(x_train,y_train)
y_pred_2=model2.predict(x_test)
r2_2=r2_score(y_test,y_pred_2)
rmse_2=np.sqrt(mean_squared_error(y_test,y_pred_2))

print("Decision Tree R2 :", r2_2)
print("Decision Tree RMSE :", rmse_2)
```

```python
model3=RandomForestRegressor(random_state=10)
model3.fit(x_train,y_train)
y_pred_3=model3.predict(x_test)
r2_3=r2_score(y_test,y_pred_3)
rmse_3=np.sqrt(mean_squared_error(y_test,y_pred_3))

print("Random Forest R2 :", r2_2)
print("Random Forest RMSE :", rmse_2)
```

```python
results = pd.DataFrame({
    "Model": ["Linear Regression", "Decision Tree", "Random Forest"],
    "R2 Score": [r2_1, r2_2, r2_3],
    "RMSE": [rmse_1, rmse_2, rmse_3]
})

print(results)
```
