```python
import pandas as pd
```

```python
df=pd.read_csv("Dataset .csv")
df
```

```python
df['Cuisines'].head()
```

```python
df["Cuisines"] = df["Cuisines"].str.split(", ")

df_exploded = df.explode("Cuisines")
```

```python
df_exploded
```

```python
avg_rating = df_exploded.groupby("Cuisines")["Aggregate rating"].mean().sort_values(ascending=False)
print("High Average Rating Cuisines : \n\n ",
      avg_rating.head(10))
```

```python
print("Low Average Rating Cuisines : \n\n",
      avg_rating.tail(10))
```

```python
votes_cuisines = df_exploded.groupby("Cuisines")["Votes"].sum().sort_values(ascending=False)

print("Most popular cuisines with high votes : \n\n", votes_cuisines.head(10))
```
