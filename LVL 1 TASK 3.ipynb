```python
import pandas as pd
```

```python
df=pd.read_csv("Dataset .csv")
df.head()
```

```python
df=df.dropna(subset=["Longitude","Latitude","Aggregate rating"])
```

```python
df
```

```python
import folium
```

```python
map_center = [df['Latitude'].mean(), df['Longitude'].mean()]
restaurant_map = folium.Map(location=map_center, zoom_start=5)

for index,row in df.iterrows():
    folium.Marker(
    location=[row['Latitude'], row['Longitude']],
    popup=f"{row['City']} - Rating: {row['Aggregate rating']}"
    ).add_to(restaurant_map)
restaurant_map.save("restaurant_map.html")
```

```python
print(restaurant_map)
```

```python
city_counts=df['City'].value_counts()
city_counts
```

```python
import matplotlib.pyplot as plt
```

```python
top10=city_counts.head(10)
plt.figure(figsize=(12,6)) 
plt.title("Top 10 cities with Restaurant counts")
plt.xlabel("City")
plt.ylabel("Number of Restaurants")
plt.bar(top10.index,top10.values,color="red")
```

```python
correlation = df['Longitude'].corr(df['Aggregate rating'])
print("Correlation between Longitude and Rating:", correlation)
```

```python
correlation = df['Latitude'].corr(df['Aggregate rating'])
print("Correlation between Latitude and Rating:", correlation)
```
